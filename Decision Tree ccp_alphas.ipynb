```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.tree import DecisionTreeClassifier

# Cargar datos
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# ==============================
# ðŸŒ³ MÃ‰TODO 1: PRUNING MANUAL
# ==============================
clf_temp = DecisionTreeClassifier(random_state=0)
path = clf_temp.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = path.ccp_alphas[:-1]  # Excluir el Ãºltimo alpha (poda total)

# ValidaciÃ³n cruzada para cada alpha
alpha_scores = []
for alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=alpha)
    scores = cross_val_score(clf, X_train, y_train, cv=5)
    alpha_scores.append([alpha, np.mean(scores), np.std(scores)])

df_alpha = pd.DataFrame(alpha_scores, columns=["alpha", "mean_accuracy", "std_accuracy"])

# GrÃ¡fico
plt.figure(figsize=(10, 6))
plt.errorbar(df_alpha["alpha"], df_alpha["mean_accuracy"], yerr=df_alpha["std_accuracy"],
             marker='o', linestyle='--', capsize=4)
plt.xlabel("ccp_alpha")
plt.ylabel("Mean Accuracy (Â± std)")
plt.title("Accuracy vs ccp_alpha (Manual Pruning)")
plt.grid(True)
plt.show()

# Mejor alpha segÃºn precisiÃ³n promedio
best_row_manual = df_alpha.loc[df_alpha["mean_accuracy"].idxmax()]
best_alpha_manual = best_row_manual["alpha"]
clf_best_manual = DecisionTreeClassifier(random_state=0, ccp_alpha=best_alpha_manual)
clf_best_manual.fit(X_train, y_train)
acc_manual = clf_best_manual.score(X_test, y_test)

print(f"ðŸŒ³ Mejor alpha (manual): {best_alpha_manual:.5f}")
print(f"ðŸŒ³ Test accuracy (manual pruning): {acc_manual:.4f}")

# ==============================
# ðŸ” MÃ‰TODO 2: GRIDSEARCHCV
# ==============================
param_grid = {'ccp_alpha': ccp_alphas}
grid = GridSearchCV(DecisionTreeClassifier(random_state=0), param_grid=param_grid, cv=5)
grid.fit(X_train, y_train)

best_alpha_grid = grid.best_params_['ccp_alpha']
clf_best_grid = grid.best_estimator_
acc_grid = clf_best_grid.score(X_test, y_test)

print(f"\nðŸ” Mejor alpha (GridSearchCV): {best_alpha_grid:.5f}")
print(f"ðŸ” Test accuracy (GridSearchCV): {acc_grid:.4f}")

# ==============================
# ðŸ“Š COMPARACIÃ“N FINAL
# ==============================
print("\nðŸ“Š COMPARACIÃ“N FINAL:")
print(f"ðŸŽ¯ Accuracy - Manual Pruning: {acc_manual:.4f}")
print(f"ðŸŽ¯ Accuracy - GridSearchCV:   {acc_grid:.4f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.model_selection import cross_val_score

# ðŸŒ³ ClasificaciÃ³n
def find_best_alpha_decision_tree(X_train, y_train, std_threshold=0.03, cv=5, random_state=0):
    clf_temp = DecisionTreeClassifier(random_state=random_state)
    path = clf_temp.cost_complexity_pruning_path(X_train, y_train)
    ccp_alphas = path.ccp_alphas[:-1]

    alpha_loop_values = []
    for ccp_alpha in ccp_alphas:
        clf = DecisionTreeClassifier(random_state=random_state, ccp_alpha=ccp_alpha)
        scores = cross_val_score(clf, X_train, y_train, cv=cv)
        alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])

    results = pd.DataFrame(alpha_loop_values, columns=['alpha', 'mean_accuracy', 'std_accuracy'])

    plt.figure(figsize=(10, 6))
    plt.errorbar(results['alpha'], results['mean_accuracy'], yerr=results['std_accuracy'],
                 marker='o', linestyle='--', capsize=4)
    plt.xlabel('ccp_alpha')
    plt.ylabel('Mean Accuracy (Â± std)')
    plt.title('Accuracy vs ccp_alpha')
    plt.grid(True)
    plt.show()

    filtered = results[results['std_accuracy'] < std_threshold]
    if filtered.empty:
        print("âš ï¸ No se encontraron valores con std < ", std_threshold)
        return None, results

    best_row = filtered.loc[filtered['mean_accuracy'].idxmax()]
    best_alpha = best_row['alpha']

    best_model = DecisionTreeClassifier(random_state=random_state, ccp_alpha=best_alpha)
    best_model.fit(X_train, y_train)

    print(f"âœ… Mejor alpha encontrado: {best_alpha:.5f}")
    print(f"ðŸŽ¯ Accuracy promedio (CV): {best_row['mean_accuracy']:.4f} Â± {best_row['std_accuracy']:.4f}")

    return best_model, results

# ðŸ“ˆ RegresiÃ³n
def find_best_alpha_regression_tree(X_train, y_train, std_threshold=0.03, cv=5, random_state=0, scoring='r2'):
    reg_temp = DecisionTreeRegressor(random_state=random_state)
    path = reg_temp.cost_complexity_pruning_path(X_train, y_train)
    ccp_alphas = path.ccp_alphas[:-1]

    alpha_loop_values = []
    for ccp_alpha in ccp_alphas:
        reg = DecisionTreeRegressor(random_state=random_state, ccp_alpha=ccp_alpha)
        scores = cross_val_score(reg, X_train, y_train, cv=cv, scoring=scoring)
        alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])

    results = pd.DataFrame(alpha_loop_values, columns=['alpha', 'mean_score', 'std_score'])

    plt.figure(figsize=(10, 6))
    plt.errorbar(results['alpha'], results['mean_score'], yerr=results['std_score'],
                 marker='o', linestyle='--', capsize=4)
    plt.xlabel('ccp_alpha')
    plt.ylabel(f'Mean {scoring} (Â± std)')
    plt.title(f'{scoring.upper()} vs ccp_alpha')
    plt.grid(True)
    plt.show()

    filtered = results[results['std_score'] < std_threshold]
    if filtered.empty:
        print("âš ï¸ No se encontraron valores con std < ", std_threshold)
        return None, results

    best_row = filtered.loc[filtered['mean_score'].idxmax()]
    best_alpha = best_row['alpha']

    best_model = DecisionTreeRegressor(random_state=random_state, ccp_alpha=best_alpha)
    best_model.fit(X_train, y_train)

    print(f"âœ… Mejor alpha encontrado: {best_alpha:.5f}")
    print(f"ðŸŽ¯ Score promedio (CV): {best_row['mean_score']:.4f} Â± {best_row['std_score']:.4f}")

    return best_model, results
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# ðŸ“¦ Cargar dataset
X, y = load_iris(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# ðŸŒ³ Obtener alphas desde pruning path para usar en GridSearch
clf_temp = DecisionTreeClassifier(random_state=0)
path = clf_temp.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas = np.unique(path.ccp_alphas[:-1])  # Quitar el Ãºltimo alpha que deja el Ã¡rbol vacÃ­o

# ðŸ” Definir hiperparÃ¡metros para buscar
param_grid = {
    'ccp_alpha': ccp_alphas,
    'max_depth': [None, 3, 5, 10],
    'min_samples_split': [2, 5, 10]
}

# ðŸ§  Inicializar modelo
clf = DecisionTreeClassifier(random_state=0)

# ðŸ”„ GridSearchCV
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(X_train, y_train)

# ðŸ“ˆ Resultados
print("ðŸ” Mejor combinaciÃ³n de hiperparÃ¡metros encontrada:")
print(grid_search.best_params_)

print(f"ðŸŽ¯ PrecisiÃ³n en test: {grid_search.best_estimator_.score(X_test, y_test):.4f}")
```
