```python
import pandas as pd

# URL base de los archivos
url_base = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/master/datasets/distributed-data/"

# Lista para almacenar los DataFrames individuales
lista_df = []

# Cargar archivos numerados del 1 al 333
for i in range(1, 333):
    nombre_archivo = f"{url_base}{i:03}.csv"  # Construye la URL completa del archivo, e.g., "https://.../001.csv"
    df = pd.read_csv(nombre_archivo)  # Lee el archivo CSV directamente desde GitHub
    lista_df.append(df)  # Agrega el DataFrame a la lista

# Combinar todos los DataFrames en uno solo
df_combinado = pd.concat(lista_df, ignore_index=True)

# Verificar el DataFrame combinado
print(df_combinado.head())
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/master/datasets/athletes/Medals.csv"
df_atleta = pd.read_csv(url, encoding = "ISO-8859-1")
df_atleta.head()
```

```python
df_atleta.shape
```

```python
url = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/master/datasets/athletes/Athelete_Country_Map.csv"
df_country = pd.read_csv(url, encoding = "ISO-8859-1" )
df_country.head()
```

```python
df_country.shape
```

```python
df_country_dp = df_country.drop_duplicates(subset="Athlete")
df_country_dp.shape
```

```python
df_sport = pd.read_csv("https://raw.githubusercontent.com/Afgcc132/python-ml-course/master/datasets/athletes/Athelete_Sports_Map.csv", encoding = "ISO-8859-1" )
df_sport.head()
```

```python
df_atleta_country = pd.merge(left=df_atleta, right=df_country_dp,
                             left_on="Athlete", right_on="Athlete")
df_atleta_country.head()
```

```python
df_atleta_country.shape
```

```python
df_atleta_country[df_atleta_country["Athlete"] == "Aleksandar Ciric"]
```

```python
df_sport_dp = df_sport.drop_duplicates(subset="Athlete")
df_sport_dp.shape
```

```python
df_atleta_country_sport = pd.merge(left=df_atleta_country, right=df_sport_dp,
                                   left_on="Athlete", right_on="Athlete")
df_atleta_country_sport.head()
```

```python
df_atleta_country_sport.shape
```

```python
df_final = df_atleta_country_sport
```

```python
df_final.head()
```

```python
df_final.isnull().sum()
```

```python
# Filtrar las filas que tienen valores nulos
filas_con_null = df_final[df_final.isnull().any(axis=1)]

print(filas_con_null)
```
