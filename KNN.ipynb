# K-Means


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing, neighbors
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split

```

```python
url = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/refs/heads/master/datasets/cancer/breast-cancer-wisconsin.data.txt"
df = pd.read_csv(url, header=None)
df.head()
```

```python
df.describe()
```

```python
df.columns = ["name", "V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9", "class"]
df.head()
```

```python
df= df.drop(["name"], axis= 1)
df.head()
```

```python
df.replace("?", -99999, inplace=True)
```

```python
Y = df['class']
X = df[["V1", "V2", "V3", "V4", "V5", "V6", "V7", "V8", "V9"]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, Y_train)
```

```python
print(X_train.shape)  # This should be (number_of_samples, 9)
```

```python
print(X_test.shape)  # This should be (number_of_samples, 9)
```

```python
accuracy = clf.score(X_test, Y_test)
print(accuracy)
```
