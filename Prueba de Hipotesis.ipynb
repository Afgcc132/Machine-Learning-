```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos
data1 = np.array([90, 98, 97, 96, 99, 98])
data2 = np.array([91, 95, 95, 99, 98, 97])

# Medias
mean1 = np.mean(data1)
mean2 = np.mean(data2)

# Gráfico
plt.figure(figsize=(8,5))
plt.scatter(np.ones(len(data1)), data1, color='blue', label='Grupo 1')
plt.scatter(np.ones(len(data2)) * 2, data2, color='red', label='Grupo 2')

# Agregar medias a la gráfica
plt.axhline(mean1, color='blue', linestyle='--', label=f'Media Grupo 1: {mean1:.2f}')
plt.axhline(mean2, color='red', linestyle='--', label=f'Media Grupo 2: {mean2:.2f}')

# Etiquetas
plt.xticks([1, 2], ['Grupo 1', 'Grupo 2'])
plt.ylabel("Valores")
plt.title("Comparación de Datos y Medias")
plt.legend()
plt.show()

# Aplicar t-test
t_stat, p_value = stats.ttest_ind(data1, data2)

print(f"T-statistic: {t_stat:.4f}, P-value: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("La diferencia es significativa (Rechazamos H0).")
else:
    print("No hay evidencia suficiente para decir que las medias son diferentes (No rechazamos H0).")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulamos dos distribuciones con diferencias pequeñas y grandes
np.random.seed(42)
group1_small = np.random.normal(50, 5, 100)
group2_small = np.random.normal(52, 5, 100)  # Medias cercanas

group1_large = np.random.normal(50, 5, 100)
group2_large = np.random.normal(70, 5, 100)  # Medias muy separadas

# Gráficos
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Caso 1: Medias cercanas
sns.kdeplot(group1_small, fill=True, alpha=0.5, label="Grupo 1", ax=axes[0])
sns.kdeplot(group2_small, fill=True, alpha=0.5, label="Grupo 2", ax=axes[0])
axes[0].set_title("Distribuciones con medias cercanas")

# Caso 2: Medias alejadas
sns.kdeplot(group1_large, fill=True, alpha=0.5, label="Grupo 1", ax=axes[1])
sns.kdeplot(group2_large, fill=True, alpha=0.5, label="Grupo 2", ax=axes[1])
axes[1].set_title("Distribuciones con medias separadas")

plt.legend()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parámetros
df = 10  # grados de libertad
x = np.linspace(-4, 4, 1000)

# Distribución normal
normal_dist = stats.norm.pdf(x, 0, 1)

# Distribución t con 10 grados de libertad
t_dist = stats.t.pdf(x, df)

# Graficamos ambas distribuciones
plt.figure(figsize=(8, 6))
plt.plot(x, normal_dist, label="Distribución Normal", color="blue", linestyle='--')
plt.plot(x, t_dist, label=f"Distribución t (df={df})", color="red")
plt.title("Comparación entre Distribución Normal y Distribución t")
plt.xlabel("Valor t")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```
