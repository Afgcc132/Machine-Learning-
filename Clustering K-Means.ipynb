```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/refs/heads/master/datasets/movies/movies.csv"
df = pd.read_csv(url, sep=";")
df.head()
```

```python
movies = df.columns.values.tolist()[1:]
movies
```

```python
from scipy.spatial.distance import cdist
import numpy as np

# Manhattan distance (p=1)
distancia1 = cdist(df[movies], df[movies], metric='minkowski', p=1)

# Euclidean distance (p=2)
distancia2 = cdist(df[movies], df[movies], metric='minkowski', p=2)

# Minkowski distance with p=10
distancia10 = cdist(df[movies], df[movies], metric='minkowski', p=10)
```

```python
def dd_to_df (dd, col_name):
  df = pd.DataFrame(dd, columns=col_name, index=col_name)

  return df
```

```python
dd_to_df(distancia1, df['user_id'])
```

```python
dd_to_df(distancia2, df['user_id'])
```

```python
dd_to_df(distancia10, df['user_id'])
```

Clustering Jer√°rquico

```python
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
```

```python
df
```

```python
movies = df.columns.values.tolist()[1:]
movies
```

```python
Z = linkage(df[movies], 'ward')
Z
```

```python
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,
    leaf_font_size=8.,
)
```

# Clustering con Python

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/Afgcc132/python-ml-course/refs/heads/master/datasets/wine/winequality-red.csv"
df = pd.read_csv(url, sep=";")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df['quality'])
```

```python
df.groupby('quality').mean()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Suponiendo que tienes un DataFrame llamado 'df'
scaler = MinMaxScaler()
df_normalizado = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

print(df_normalizado)
```

```python
df_normalizado.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
clus = AgglomerativeClustering(n_clusters=6, linkage="ward").fit(df_normalizado)
clus.labels_
```

```python
md= pd.Series(clus.labels_)
md.value_
```

```python
plt.hist(md)
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage
import matplotlib.pyplot as plt
```

```python
Z = linkage(df_normalizado, 'ward')
Z
```

```python
plt.figure(figsize=(25,10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(
    Z,
    leaf_rotation=90.,
    leaf_font_size=8.,
    truncate_mode='lastp',
    show_leaf_counts=True,
    show_contracted=True,
)
```

K- means


```python
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
model = KMeans(n_clusters=6)
model.fit(df_normalizado)
```

```python
md_k = pd.Series(model.labels_)
md_k.value_counts()
```

```python
plt.hist(md_k)
```

```python
df_normalizado['clust_k']= md_k
df_normalizado['clust_h']= md
```

```python
df_normalizado.head()
```

```python
df_normalizado.groupby('clust_k').mean()
```
